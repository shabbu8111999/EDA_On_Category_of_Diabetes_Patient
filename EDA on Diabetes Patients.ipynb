# Cleaning the Data

```python
# Checking the null values

data.isnull().sum()
```

```python
Patient = data.drop(['Outcome','SkinThickness','BloodPressure','Pregnancies'], axis = 1)
```

```python
Patient.head()
```

# Analysis of the Data

```python
# Performing Correlation Analysis

Correlation = Patient.corr()
```

```python
sns.heatmap(Correlation, xticklabels = Patient.columns, yticklabels = Patient.columns, annot = True)
```

```python
sns.pairplot(Patient) # It will calculate each relations in data.
```

```python
# Relation plot

sns.relplot(x = 'DiabetesPedigreeFunction', y = 'BMI', hue = 'Age', data = Patient)
```

```python
# Plotting The Histogram plot 

sns.histplot(Patient['DiabetesPedigreeFunction'])
```

```python
# Let's plot the Distribution using Seaborn

sns.distplot(Patient['BMI'], bins = 10)
```

```python
sns.histplot(Patient['Insulin'])
```

```python
sns.distplot(Patient['Glucose'], bins = 10)
```

```python
# Let's check the categorical plot

sns.catplot(x = 'BMI', kind = 'box', data = Patient)
```

```python
sns.catplot(x = 'Insulin', kind = 'box', data = Patient)
```
